```python
import time
import numpy as np
from numpy import random
from time import time_ns
a=np.random.randint(1,10000,100)
print(a)
```

```python
#bubble sort
b=a
start=time.time_ns()
for i in range(len(b)):
    for j in range(len(b)-i-1):
        if(a[j]>a[j+1]):
            b[j],b[j+1]=b[j+1],b[j]
final=time.time_ns()
x=final-start
print(x)
print(b)
```

```python
# insertion sort
c=a
start=time.time_ns()
for i in range(1,len(c)):
    key=c[i]
    j=i-1
    while key < c[j] and j>=0:
        c[j+1]=c[j]
        j=j-1
        c[j+1]=key
final=time.time_ns()
y=final-start
print(y)
print(c)
```

```python
#selection sort
start=time.time_ns()
d=a
for i in range(len(d)):
    min=i
    for j in range(i+1,len(d)):
        if d[min]>d[j]:
            d[min],d[j]=d[j],d[min]
final=time.time_ns()
z=final-start
print(z)
print(d)
```

```python
import matplotlib.pyplot as plt
plt.scatter(len(a),x,color='blue')
plt.scatter(len(a),y,color='green')
plt.scatter(len(a),z,color='red')
plt.ylabel('Time Taken')
plt.xlabel('Sortings')
plt.show()
```
